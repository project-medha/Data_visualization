# Outliers

The code demonstrates how the outliers skew up the conventional averaging metrics and representations. Hence the techniques that can be employed for better data visualization.

```python
%matplotlib inline
import numpy as np

incomes = np.random.normal(27000, 15000, 10000)
#normal distribution obtained with mean = 27000, standard deviation= 15000 
#and no of data points = 10000
incomes = np.append(incomes, [1000000000])
#appending an outlier in the incomes array 
import matplotlib.pyplot as plt
plt.hist(incomes, 50)#initiating parameter: array and bins
plt.show()
```

Impact of Outliers on data:\
 One billionaire ended up squeezing everybody else into a single line in the histogram. Plus it skewed the mean income significantly.

```python
incomes.mean()
```

Filtering Principle for outliers: \
If the datapoint lies outside the scope of twice the standard deviation from median it is an outlier 

```python
#code snippet containing the function to reject outliers in a principled manner 

def reject_outliers(data):
    u = np.median(data)
    s = np.std(data)
    filtered = [e for e in data if (u - 2 * s < e < u + 2 * s)]
    print (u)
    print (s)
    #code to iterate through data array apply condition on each element
    #and store it as array element
    return filtered

filtered = reject_outliers(incomes)

plt.hist(filtered, 50)
plt.show()
```

```python
np.mean(filtered)
```

```python
def filtering_2(data):
    m2=np.median(data)
    st2=np.std(data)
    filtered_data=[e for e in data if (m2-3*st2 < e < m2 + 3 * st2)]
    print(m2)
    print(st2)
    return (filtered_data)
```

```python
out_liers=np.random.randint(10000000,1000000000,10)
print(out_liers)
```

```python
incomes = np.append(incomes,out_liers)
```
