Pair plot  from Seaborn. \
It lets us visualize plots of every combination of various attributes together, so we can look for interesting patterns between features.

```python
df2 = df[['Cylinders', 'CityMPG', 'HwyMPG', 'CombMPG']]
df2.head()
```

```python
sns.pairplot(df2, height=2.5); 
```

Conclusions from graph above: \
 We can see there is a relationship between number of cylinders and MPG, but MPG for 4-cylinder vehicles ranges really widely.\
There also appears to be a good linear relationship between the different ways of measuring MPG values, until we get into the higher MPG ratings.

Seaborn- "scatterplot"\
It plots individual data points across two axes of our choosing, so we can see how our data is distributed across those dimensions.

```python
sns.scatterplot(x="Eng Displ", y="CombMPG", data=df)
```

Seaborn - "jointplot" \
It combines a scatterplot with histograms on both axes. It lets us visualize both the individual data points and the distribution across both dimensions at the same time. 

```python
sns.jointplot(x="Eng Displ", y="CombMPG", data=df)
```

Seaborn - "lmplot" \
It is a scatterplot, but with a linear regression line computed and overlaid onto the data.

```python
sns.lmplot(x="Eng Displ", y="CombMPG", data=df)
```

Seaborn - "box plot." \
 It is also called a "box and whiskers" plot, which is useful for visualizing typical values for a given category without getting distracted by outliers. Each box represents the range between the first and third quartiles of the data, with a line representing the median value. The "whiskers" that extend from the box represent the spread of the remainder of the data, apart from clear outliers that are plotted as individual points outside of the whiskers.


```python
ax=sns.boxplot(x='Mfr Name', y='CombMPG', data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```

Swarm Plot \
 Instead of boxes and whiskers, it plots each individual data point - but does so in such way that groups them together based on their distribution. 

```python
ax=sns.swarmplot(x='Mfr Name', y='CombMPG', data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```

Countplot \
 This is basically the same thing as a histogram, but for categorical data. \
 It lets you count up how many times each given category on the X axis occurs in your data, and plot it. 

```python
ax=sns.countplot(x='Mfr Name', data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```

Seaborn - "Heatmap" \
A heat map allows us to plot tabular, 2D data of some sort, with colors representing the individual values in each cell of the 2D table.



```python
df2 = df.pivot_table(index='Cylinders', columns='Eng Displ', values='CombMPG', aggfunc='mean')
print(df2)
sns.heatmap(df2)
```

Seaborn- "scatterplot" \
It arranges itself into columns when we have ordinal data like the number of gears.

```python
sns.scatterplot(x='# Gears', y="CombMPG", data=df)
```

Seaborn- "lmplot" \
It gives us a linear regression of our data overlaid on the graph,

```python
sns.lmplot(x='# Gears', y="CombMPG", data=df)
```

Seaborn - "jointplot"\
It gives histogram on each axis. 

```python
sns.jointplot(x='# Gears', y="CombMPG", data=df)
```

Conclusion:The most common gear configuration seems to be 8, with 6 closely behind. And MPG ratings seem to roughly follow a bell curve centered at around 22 or so.

 Seaborn - "boxplot" \
 It shows us many of the extreme values are in fact outliers that are best discarded when analyzing the trends.
 

```python
sns.boxplot(x='# Gears', y="CombMPG", data=df)
```

Conclusion: Those continuously variable transmissions with a single gear are really quite good at fuel efficiency (higher MPG's are more efficient). Between 5 and 8 things are roughly the same, but from 8-10 things decline markedly.

Seaborn - "swarmplot"

```python
sns.swarmplot(x='# Gears', y="CombMPG", data=df)
```

Conclusion:It is apparent that those high MPG outliers are in fact outliers on the 6-gear vehicles; the vast majority of 6-gear vehcles have less than 32 MPG.

Our final verdict: \
More gears result in worse fuel efficiency.
